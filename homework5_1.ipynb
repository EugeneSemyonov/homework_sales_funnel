```python
import csv
# name.csv ниже нужно заменить на название csv файла, который вы хотите открыть
file_path = '..\homework5\click_stream3.csv'
#%pwd
''' Для удобства анализа я написал функцию выводящую воронку построчно. 
Мог бы написать функцию которая генерирует html-документ с таблицей, но пожалел времени, т.к. данных немного и считаю, что 
мне удалось успешно проанализировать данные без сравнения в таблице.
'''
def print_funnel(funnel):
    for key, value in funnel.items():
        if isinstance(value, dict):
            print('\n'+str(key))
            print_funnel(value)
        else:
            print(key, value)
   

with open(file_path, mode='r') as csv_file: # открываем файл
    csv_reader = csv.DictReader(csv_file, fieldnames=['ID','page','date']) # читаем файл
    
    final_funnel = {'Male':{'Desktop':{}, 'Mobile':{}}, 'Female':{'Desktop':{}, 'Mobile':{}}}
    funnel_template = {'homepage_users':0, 'search_users':0, 'payment_page_users':0, 'paying_users':0}
    for row in csv_reader: #перебираем по одной строчке нашего файла
        page = list(row.items())[1][1] #так можно получить первый (не нулевой) элемент строки -
        # для нашего датасета это строка, указывающая, на какой страничке было совершено действие

        # мой код для расчета воронки:
#       date = list(row.items())[2][1]
        number_of_month = int(list(row.items())[2][1][5:7])
        device = list(row.items())[3][1][0]
        gender = list(row.items())[3][1][1]

        if page == '1_home_page':
            current_month_funnel = final_funnel[gender][device].setdefault(number_of_month, funnel_template.copy())
            current_month_funnel['homepage_users'] += 1
        elif page == '2_search_page':
            current_month_funnel = final_funnel[gender][device].setdefault(number_of_month, funnel_template.copy())
            current_month_funnel['search_users'] += 1
        elif page == '3_payment_page':
            current_month_funnel = final_funnel[gender][device].setdefault(number_of_month, funnel_template.copy())
            current_month_funnel['payment_page_users'] += 1
        elif page == '4_payment_confirmation_page':
            current_month_funnel = final_funnel[gender][device].setdefault(number_of_month, funnel_template.copy())
            current_month_funnel['paying_users'] += 1
        
    print_funnel(final_funnel)

#    print(final_funnel['Male'])
#    print(final_funnel['Female'])
#     print("Зашли на главную страницу:", homepage_users)
#     print("Зашли на страницу поиска:", search_users)
#     print("Зашли на страницу оплаты:", payment_page_users)
#     print("Совершили оплату заказа:", paying_users)
            
```
